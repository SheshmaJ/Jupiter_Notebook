```python
pivot = pd.pivot_table(df, index=['Payment Method'],values=['Previous Purchases','Review Rating'])
pivot
```

```python
import numpy as np
```

```python
pivot = pd.pivot_table(df, index=['Payment Method'],values=['Previous Purchases','Review Rating'],aggfunc=np.sum)
pivot
```

```python
pivot = pd.pivot_table(df, index=['Payment Method','Category'],values=['Previous Purchases','Review Rating'],aggfunc={'Previous Purchases':np.sum,'Review Rating':np.mean})
pivot
```

```python
pivot.plot()
```

```python
pivot.plot(kind='bar',title='purchase vs review')
```
